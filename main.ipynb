```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import time
```

```python
def fouier_series(x: np.ndarray, n: int) -> np.ndarray:
    """フ―リエ級数を計算する

    Args:
        x (np.ndarray): 入力値
        n (int): 項数
    
    Returns:
        np.ndarray: フ―リエ級数の値
    """
    # フ―リエ級数の値を計算する
    y = np.zeros_like(x)
    for i in range(1, n+1):
        y += 2 * (-1) ** (i+1) / i * np.sin(i * x)
    return y
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
N = [1, 3, 5]
y = [fouier_series(x, n) for n in N]
ux = np.linspace(-np.pi, np.pi, 1000)
```

```python
plt.style.use(['ieee', 'grid'])
plt.figure(figsize=(7, 6))

plt.plot(x, ux, label="u(x)=x")

for i in range(len(N)):
    plt.plot(x, y[i], label=f"N={N[i]}")

plt.title("Fourier series expansion\n$u(x)=x$, $x \in [-\pi, \pi]$\n$u(x) \simeq 2 \sum_{n=1}^{N} \\frac{(-1)^{n+1}}{n} \sin(nx)$\n")
plt.xlabel("x")
plt.ylabel("y")
# πの目盛りを表示する
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r"$-\pi$", r"$-\frac{\pi}{2}$", r"$0$", r"$\frac{\pi}{2}$", r"$\pi$"])
# y = 0を実践軸に表示する
# plt.axhline(y=0, color='black', linewidth=0.5)
plt.legend()
# タイトルと図の間に余白を設ける
plt.tight_layout()
now_time = time.strftime("%Y%m%d%H%M%S")
plt.savefig(f"fourier_series_{now_time}.png")
plt.show()
```
